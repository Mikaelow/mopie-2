```python
import pandas as pd
duplex_level = pd.read_excel('PlanDwupoziomowy.xlsx')
duplex_level
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def model_equation(x, a0, a1, a2):
    return a0 + a1 * x['x1'] + a2 * x['x2'] 

initial_guess = [0, 0, 0]
coefficients, covariance = curve_fit(model_equation, duplex_level[['x1', 'x2']], duplex_level['Value'], p0=initial_guess)
coefficients
```

#### Basic

##### Formating data for plot

```python
import numpy as np

a0 = coefficients[0]
a1 = coefficients[1]
a2 = coefficients[2]

x1 = duplex_level['x1']
x2 = duplex_level['x2']
x1_value = pd.DataFrame(np.linspace(min(x1),max(x1),100),columns=['x1'])
x2_value = pd.DataFrame(np.linspace(min(x2),max(x2),100),columns=['x2'])
xyz_values = pd.merge(x1_value,x2_value,how='cross')
z_value = model_equation(xyz_values, a0, a1, a2)
xyz_values['Value'] = z_value
xyz_values
```

##### Plot

```python
import plotly.graph_objects as go
import plotly.express as px

x1 = xyz_values['x1'].tolist()
x2 = xyz_values['x2'].tolist()
z = xyz_values['Value'].tolist()

# Tworzenie wykresu 3D jako powierzchni
fig = px.scatter_3d(xyz_values, x='x1', y='x2', z='Value', color='Value', title='Wykres 3D', opacity=0.8, size_max=10)
fig.show() 
```
