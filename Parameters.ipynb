### Create data base

```python
import pandas as pd
import io

db = pd.read_excel('dane.xlsx')

```

### Parameters


#### Mean

```python
###t1
t1_minus = db[db['t1']=='t1-']
t1_mean_minus = t1_minus['Value'].agg('mean')
t1_plus = db[db['t1']=='t1+']
t1_mean_plus = t1_plus['Value'].agg('mean')
t1_mean = [t1_mean_minus,t1_mean_plus]
diff_mean1 = abs(t1_mean[0]-t1_mean[1])

###t2
t2_minus = db[db['t2']=='t2-']
t2_mean_minus = t2_minus['Value'].agg('mean')
t2_plus = db[db['t2']=='t2+']
t2_mean_plus = t2_plus['Value'].agg('mean')
t2_mean = [t2_mean_minus,t2_mean_plus]
diff_mean2 = abs(t2_mean[0]-t2_mean[1])

###t3
t3_minus = db[db['t3']=='t3-']
t3_mean_minus = t3_minus['Value'].agg('mean')
t3_plus = db[db['t3']=='t3+']
t3_mean_plus = t3_plus['Value'].agg('mean')
t3_mean = [t3_mean_minus,t3_mean_plus]
diff_mean3 = abs(t3_mean[0]-t3_mean[1])

###t4
t4_minus = db[db['t4']=='t4-']
t4_mean_minus = t4_minus['Value'].agg('mean')
t4_plus = db[db['t4']=='t4+']
t4_mean_plus = t4_plus['Value'].agg('mean')
t4_mean = [t4_mean_minus,t4_mean_plus]
diff_mean4 = abs(t4_mean[0]-t4_mean[1])

dif=[diff_mean1,diff_mean2,diff_mean3,diff_mean4]
print(f'All difference mean {dif}')
```

### Bilans scatter

```python
import matplotlib.pyplot as plt
import matplotlib.lines as lines


fig, ax = plt.subplots()

ax.scatter(t1_minus['t1'],t1_minus['Value'],c='b')
ax.scatter(t1_plus['t1'],t1_minus['Value'],c='b')

ax.scatter(t2_minus['t2'],t2_minus['Value'],c='orange')
ax.scatter(t2_plus['t2'],t2_plus['Value'],c='orange')

ax.scatter(t3_minus['t3'],t3_minus['Value'],c='red')
ax.scatter(t3_plus['t3'],t3_plus['Value'],c='red')


ax.scatter(t4_minus['t4'],t4_minus['Value'],c='purple')
ax.scatter(t4_plus['t4'],t4_plus['Value'],c='purple')
plt.ylabel('Napięcie U [V]')
plt.xlabel('Kombinacja skrajnych zmiennej standaryzowanych')
plt.grid(True)
```
