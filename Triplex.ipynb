### Triplex Level

#### Basic

##### Read data from Triplex_data.xlsx

```python
import pandas as pd
triplex_level = pd.read_excel('Triplex_data.xlsx')
triplex_level = triplex_level[['t2','t3','x2','x3','Value']].rename(columns={'x2':'x1', 'x3':'x2'})
triplex_level
```

##### Creating function

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def model_equation(x, a0, a1, a2, a3, a4, a5):
    return a0 + a1 * x['x1'] + a2 * x['x2'] + a3 * pow(x['x1'],2) + a4 * pow(x['x2'],2) + a5 * x['x1'] * x['x2']

initial_guess = [0, 0, 0, 0, 0, 0]
coefficients, covariance = curve_fit(model_equation, triplex_level[['x1', 'x2']], triplex_level['Value'], p0=initial_guess)
coefficients
```

##### Formating data for plot

```python
a0=coefficients[0]
a1=coefficients[1]
a2=coefficients[2]
a3=coefficients[3]
a4=coefficients[4]
a5=coefficients[5]

x1 = triplex_level['x1']
x2 = triplex_level['x2']
x1_value = pd.DataFrame(np.linspace(min(x1),max(x1),100),columns=['x1'])
x2_value = pd.DataFrame(np.linspace(min(x2),max(x2),100),columns=['x2'])
xyz_values = pd.merge(x1_value,x2_value,how='cross')
z_value = model_equation(xyz_values, a0, a1, a2, a3, a4, a5)
xyz_values['Value'] = z_value
xyz_values
```

##### Plot

```python
import plotly.graph_objects as go
import plotly.express as px
fig = px.scatter_3d(xyz_values, x='x1', y='x2', z='Value', color='Value', title='Wykres 3D', opacity=0.8, size_max=10)
fig.show() 
```
